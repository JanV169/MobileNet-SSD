```python
# deleting anime with 0 rating
anime_df=anime_df[~np.isnan(anime_df["rating"])]

# filling mode value for genre and type
anime_df['genre'] = anime_df['genre'].fillna(
anime_df['genre'].dropna().mode().values[0])

anime_df['type'] = anime_df['type'].fillna(
anime_df['type'].dropna().mode().values[0])

#checking if all null values are filled
anime_df.isnull().sum()
```

```python
rating_df['rating'] = rating_df['rating'].apply(lambda x: np.nan if x==-1 else x)
rating_df.head(20)
```

```python
# rearranging and restructuring the ds for the final processing
anime_df = anime_df[anime_df['type']=='TV']

# final_df -> rated_anime_df
rated_anime_df = rating_df.merge(anime_df, left_on = 'anime_id', right_on = 'anime_id', suffixes= ['_user', ''])
rated_anime_df =rated_anime_df[['user_id', 'name', 'rating']]
rated_anime_df= rated_anime_df[rated_anime_df.user_id <= 7500]
rated_anime_df.head()
```

-data engineering

```python
pivot = rated_anime_df.pivot(index=['user_id'], columns=['name'], values='rating')
pivot.head()
```

```python
pivot_n = pivot.apply(lambda x: (x-np.mean(x))/(np.max(x)-np.min(x)), axis=1) # normalization
pivot_n.fillna(0, inplace=True) # filling NaN values as 0
pivot_n = pivot_n.T # transposing the pivot
pivot_n = pivot_n.loc[:, (pivot_n != 0).any(axis=0)] # removing unrated values
piv_sparse = sp.sparse.csr_matrix(pivot_n.values)
```

-cosine similarity model

```python
#model based on anime similarity
anime_sim = cosine_similarity(piv_sparse)

#Df of anime similarities
anime_sim_df = pd.DataFrame(anime_sim, index = pivot_n.index, columns = pivot_n.index)
```

```python
def anime_recom(anime_name):
    """
    This function will return the top 5 shows with the highest cosine similarity value and show match percent
    
    example:
    >>>Input: 
    
    anime_recom('Death Note')
    
    >>>Output: 
    
    Recommended because you watched Death Note:

                    #1: Code Geass: Hangyaku no Lelouch, 57.35% match
                    #2: Code Geass: Hangyaku no Lelouch R2, 54.81% match
                    #3: Fullmetal Alchemist, 51.07% match
                    #4: Shingeki no Kyojin, 48.68% match
                    #5: Fullmetal Alchemist: Brotherhood, 45.99% match 

               
    """
    number = 1
    print('Recommended because you watched {}:\n'.format(anime_name))
    for anime in anime_sim_df.sort_values(by = anime_name, ascending = False).index[1:6]:
        print(f'#{number}: {anime}, {round(anime_sim_df[anime][anime_name]*100,2)}% match')
        number +=1  
```

```python
anime_recom('Fullmetal Alchemist')
```
